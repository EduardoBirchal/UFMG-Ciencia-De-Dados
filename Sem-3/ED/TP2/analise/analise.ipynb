```python
# -*- coding: utf-8 -*-
"""
Notebook para gerar um gráfico a partir de um arquivo CSV.
"""

# Importação das bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- Crie um arquivo CSV de exemplo ---
# Esta seção cria um arquivo 'dados.csv' para fins de demonstração.
# # Se você já possui seu arquivo 'dados.csv', pode comentar ou remover esta parte.
# dados_exemplo = {
#     'numero': [1, 2, 3, 4, 5, 6, 7, 8],
#     'operacoes': [10, 25, 15, 30, 20, 40, 35, 50]
# }
# df_exemplo = pd.DataFrame(dados_exemplo)
# nome_arquivo_csv = 'dados.csv'
# df_exemplo.to_csv(nome_arquivo_csv, index=False)

# print(f"Arquivo '{nome_arquivo_csv}' de exemplo criado com sucesso.")
# print("-" * 30)
# # -----------------------------------------
nome_arquivo_csv = 'dados.csv'
# Verifica se o arquivo CSV existe no diretório
if not os.path.exists(nome_arquivo_csv):
    print(f"Erro: O arquivo '{nome_arquivo_csv}' não foi encontrado.")
    print("Por favor, certifique-se de que o arquivo está no mesmo diretório que o notebook.")
else:
    # Carrega os dados do arquivo CSV para um DataFrame do pandas
    try:
        dados = pd.read_csv(nome_arquivo_csv)

        # Exibe as primeiras linhas do DataFrame para verificação
        print("Dados carregados do arquivo CSV:")
        print(dados.head())
        print("-" * 30)

        # Define as colunas para os eixos X e Y
        eixo_x = 'Capacidade de Transporte'
        eixo_y = 'Operacoes Totais'

        # Verifica se as colunas esperadas existem no DataFrame
        if eixo_x in dados.columns and eixo_y in dados.columns:
            # Cria o gráfico
            plt.figure(figsize=(10, 6))  # Define o tamanho da figura do gráfico

            plt.plot(dados[eixo_x], dados[eixo_y], marker='o', linestyle='-', color='b')

            # Adiciona títulos e rótulos aos eixos
            plt.title('Relação entre Capacidade do Transporte e Operações', fontsize=16)
            plt.xlabel(eixo_x.capitalize(), fontsize=12)
            plt.ylabel(eixo_y.capitalize(), fontsize=12)

            # Adiciona uma grade ao gráfico para melhor visualização
            plt.grid(True)

            # Exibe o gráfico
            plt.show()

        else:
            print(f"Erro: As colunas '{eixo_x}' e/ou '{eixo_y}' não foram encontradas no arquivo CSV.")
            print(f"Colunas encontradas: {list(dados.columns)}")

    except Exception as e:
        print(f"Ocorreu um erro ao processar o arquivo CSV: {e}")
```
